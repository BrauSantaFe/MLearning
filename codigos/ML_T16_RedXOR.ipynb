```python
import numpy as np

# Definición de pesos y bias
w1 = np.array([[0.5, 0.3], [-0.5, -0.2]])
b1 = np.array([0.1, -0.1])
w2 = np.array([[0.4], [-0.3]])
b2 = np.array([0.2])
x = np.array([[0, 0], [0, 1], [1, 1], [1, 0]]) 
y = np.array([[0], [1], [1], [0]])  
alfa = 0.5  
epochs = 10000  
```

```python
# Función de activación sigmoide
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Función para calcular la salida de una neurona
def neurona(w, b, x):
    z = np.dot(x, w) + b
    return sigmoid(z)
```

```python
# Entrenamiento 
for epoch in range(epochs):
    # Propagación hacia adelante
    h = neurona(w1, b1, x)  # Capa oculta
    y_hat = neurona(w2, b2, h)  # Capa de salida

    # Cálculo de la derivada de la pérdida respecto a z
    def dLdz(y, y_hat):
        return y_hat - y  # Derivada estándar de pérdida de entropía cruzada

    # Derivada de la capa de salida
    dL_dz = dLdz(y, y_hat)

    # Actualización de los pesos y bias de la capa de salida
    dw2 = np.dot(h.T, dL_dz)
    db2 = np.sum(dL_dz, axis=0)
    w2 -= alfa * dw2
    b2 -= alfa * db2

    # Propagación hacia atrás a la capa oculta
    dL_dh = np.dot(dL_dz, w2.T)
    dL_dz_h = dL_dh * h * (1 - h)  # Derivada de la sigmoide
    dw1 = np.dot(x.T, dL_dz_h)
    db1 = np.sum(dL_dz_h, axis=0)
    w1 -= alfa * dw1
    b1 -= alfa * db1
```

```python
# mostramos la salida original y la predicha
print("Salida original (y):")
print(y)
print("Salida predicha (y_hat):")
print(np.round(y_hat))  
```
