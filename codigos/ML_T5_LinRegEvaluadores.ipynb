```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Cargamos el dataset
data = pd.read_csv('peso_altura_sexo.csv')
print(data.head())
```

```python
#x1 es la altura
#x2 es el sexo
#y es el peso
x1 = data.iloc[:, 2].values
x2 = data.iloc[:, 1].values
y = data.iloc[:, 3].values

# Normalizamos los datos
x1 = (x1 - np.mean(x1)) / np.std(x1)
x2 = (x2 - np.mean(x2)) / np.std(x2)
y = (y - np.mean(y)) / np.std(y)
```

```python
def GradienteDescent(x1, x2, y, alpha):
    N = len(y)
    X = np.c_[np.ones(N), x1, x2]
    w = np.dot(np.linalg.inv(np.dot(X.T,X)),np.dot(X.T,y))

    return w
```

```python
def SSE(y, ypred):
    return np.sum((y - ypred)**2)
```

```python
def SST(y):
    return np.sum((y - np.mean(y))**2)
```

```python
def R2(y, ypred):
    return 1 - SSE(y, ypred) / SST(y)
```

```python
def MSR(y, ypred, p):
    return SSE(y, ypred) / (p - 1)
#normalizando la suma de los errores cuadráticos por los 
# grados de libertad asociados a los parámetros del modelo (𝑝−1)
```

```python
def MSE(y, ypred, p):
    return SSE(y, ypred) / (len(y) - p)
```

```python
def F(y, ypred, p):
    return MSR(y, ypred, p) / MSE(y, ypred, p)
```

```python
def generate_y(w, x1, x2):
    return w[0] + w[1]*x1 + w[2]*x2
```

```python
# Calculamos los pesos
alpha = 0.01
w = GradienteDescent(x1, x2, y, alpha)
print(f'w = {w}')
```

```python
# Generamos y_hat
y_hat = generate_y(w, x1, x2)
```

```python
# Calculamos el R2
r2 = R2(y, y_hat)
print(f'R2 = {r2}')
```

```python
# Calculamos el F
p = len(w) #grado de libertad
f = F(y, y_hat, p)
print(f'F = {f}')
```
